Despúes, realicé el cálculo de los servicios pendientes por cobrar, que no están incluidos dentro de los planes contratados por los usuarios.

```python
# Calcula el ingreso mensual para cada usuario
```

```python
# Unir los dos dataframe consumo_usuario_tarifa y tarifas

tarifas = tarifas.rename(columns={'plan_name': 'plan'})

ingresos = pd.merge(consumo_usuarios_tarifa, tarifas, on='plan', how='outer')

ingresos
```

```python
def por_cobrar(incluido, consumido):
    if consumido > incluido:
        return consumido - incluido
    else:
        return 0

def solo_positivos(valor):
    if valor > 0:
        return valor
    else:
        return 0

ingresos['min_por_cobrar'] = ingresos['cantidad_min'] - ingresos['minutes_included']
ingresos['min_por_cobrar'] = ingresos['min_por_cobrar'].apply(solo_positivos)

ingresos['sms_por_cobrar'] = ingresos['cantidad_SMS'] - ingresos['messages_included']
ingresos['sms_por_cobrar'] = ingresos['sms_por_cobrar'].apply(solo_positivos)

ingresos['mb_por_cobrar'] = ingresos['cantidad_mb'] - ingresos['mb_per_month_included']
ingresos['mb_por_cobrar'] = ingresos['mb_por_cobrar'].apply(solo_positivos)

ingresos.sample(10, random_state=500)
```

```python
# Cálculo del plan mensual incluyendo el cobro de minutos consumidos fuera del plan

ingresos['min_cobro_exceso'] = ingresos['min_por_cobrar'] * ingresos['usd_per_minute']

ingresos['plan_mensual_llamadas_exceso'] = ingresos['min_cobro_exceso'] + ingresos['usd_monthly_pay']

ingresos
```

## Prueba las hipótesis estadísticas

Hipótesis nula, Ho. Los ingresos de los usuarios del plan Surf son iguales a los ingresos de los usuarios del plan Ultimate.

Hipótesis alternativa, H1.Los ingresos de los usuarios del plan Surf son diferentes a los ingresos de los usuarios del plan Ultimate.

```python
# Prueba las hipótesis
```

```python
# Ingresos mensuales de plan Surf

ingresos_surf = ingresos[ingresos['plan'] == 'surf']['ingresos_mensuales']

ingresos_surf
```

```python
# Ingresos mensuales de plan Ultimate

ingresos_ultimate = ingresos[ingresos['plan'] == 'ultimate']['ingresos_mensuales']

ingresos_ultimate
```

```python
# Probar la hipótesis

alpha = 0.05

results = st.ttest_ind(ingresos_surf, ingresos_ultimate)

print('valor_p:', results.pvalue)

if results.pvalue < alpha:
    print('Rechazar la hipótesis nula, Ho')
else:
    print('No se puede rechazar la hipótesis nula, Ho')
```
