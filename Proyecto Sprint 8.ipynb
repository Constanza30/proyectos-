```python
# Tiempo de duraci√≥n de los viajes en los s√°bados no lluviosos

duration_good_saturday_trips = good_saturday_trips['duration_seconds']

duration_good_saturday_trips
```

#### **5.5. Prueba de Hip√≥tesis**

##### 5.5.1. Prueba de hip√≥tesis

La hip√≥tesisis que plantea este proyecto es una Hip√≥tesis alternativa, H1: "La duraci√≥n promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare cambia los s√°bados lluviosos", debido a que hace referencia a que hay cambios en la duraci√≥n del los viajes en los d√≠as llluviosos.

Por lo tanto palenteo como Hip√≥tesis nula, Ho: "La duraci√≥n promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare no cambia los s√°bados".

```python
# Utilizara prueba t de Students para la prueba de la hip√≥tesis nula Ho

alpha = 0.05

results = st.ttest_ind(duration_rainy_saturday_trips, duration_good_saturday_trips)

print('valor_p:', results.pvalue)

if results.pvalue < alpha:
    print('Rechazar la hip√≥tesis nula, Ho')
else:
    print('No se puede rechazar la hip√≥tesis alternativa, H1')
```
